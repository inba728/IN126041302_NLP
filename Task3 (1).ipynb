Chatbot using Transformers

This chatbot is built using GPT-2 (Generative Pre-trained Transformer 2), a powerful language model developed by OpenAI and available on the Hugging Face Model Hub.

Key Features:

Model: gpt2 (Decoder-only Transformer architecture)

Library: Hugging Face transformers pipeline

Capability: Uses Natural Language Generation (NLG) to generate meaningful responses

Design: Interactive console chatbot with controlled responses for important questions

System Architecture & Pipeline Flow

The chatbot follows a standard NLP pipeline:

User Input: Captures input using input()

Model Processing: Text is tokenized into numerical format
Transformer processes the input
Response Generation: Model predicts the best response

Display Output: Response is shown to user

Loop Until Exit: Continues until user types exit

Visual Representation:

User Input → Tokenization → Transformer Processing → Decoding → Display Output → Loop

```python
import warnings
from transformers import pipeline, set_seed

# Suppress warnings
warnings.filterwarnings("ignore")

# Set seed for consistent results
set_seed(42)

print("Libraries imported successfully.")
```

```python
print("Loading GPT-2 Model... Please wait.")

# Load GPT-2 model
generator = pipeline('text-generation', model='gpt2')

print("Model loaded and ready.")
```

```python
def get_chatbot_response(user_input):

    user_input_lower = user_input.lower().strip()

    # Basic conversational responses
    social_responses = {
        "hello": "Hello! Nice to meet you. How can I assist you today?",
        "hi": "Hi there! How can I help you today?",
        "thank you": "You're welcome! Feel free to ask more questions.",
        "thanks": "No problem! Anything else I can help with?"
    }

    # Check simple responses first
    if user_input_lower in social_responses:
        return social_responses[user_input_lower]

    # Important factual responses
    elif "python" in user_input_lower:
        return "Python was created by Guido van Rossum and released in 1991."

    elif "artificial intelligence" in user_input_lower or "what is ai" in user_input_lower:
        return "Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning and problem solving."

    # Transformer-based response
    else:
        prompt = f"User: {user_input}\nAssistant:"

        outputs = generator(
            prompt,
            max_new_tokens=30,
            num_return_sequences=1,
            pad_token_id=50256
        )

        generated_text = outputs[0]['generated_text']

        # Extract only chatbot response
        response = generated_text.split("Assistant:")[-1].strip()

        return response if response else "Tell me more!"
```
